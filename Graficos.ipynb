```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the updated sleep data
data = pd.read_csv('sleep_data(extra)1.csv', parse_dates=['Sleep_Start', 'Sleep_End'])

# Calculate sleep duration
data['Sleep_Duration'] = (data['Sleep_End'] - data['Sleep_Start']).dt.total_seconds() / 3600  # in hours

# Calculate average sleep time
data['Sleep_Start_Hour'] = data['Sleep_Start'].dt.hour + data['Sleep_Start'].dt.minute / 60

# Analyze average sleep duration
avg_sleep_duration = data['Sleep_Duration'].mean()
print(f"Duración Promedio del Sueño: {avg_sleep_duration:.2f} horas")

# Analyze average sleep start time
avg_sleep_start = data['Sleep_Start_Hour'].mean()
print(f"Tiempo Promedio de Inicio del Sueño: {avg_sleep_start:.2f} horas")

# Analyze sleep quality
avg_sleep_quality = data['Sleep_Quality'].mean()
print(f"Calificación de Calidad Promedio del Sueño: {avg_sleep_quality:.2f}")

# Analyze average heart rate during sleep
avg_heart_rate = data['Heart_Rate_Avg'].mean()
print(f"Tasa Cardíaca Promedio Durante el Sueño: {avg_heart_rate:.2f} PPM (Pulsos por minuto)")

# Plot Sleep Duration
plt.figure(figsize=(10, 5))
plt.bar(data['Date'], data['Sleep_Duration'], color='skyblue')
plt.title('Duración del sueño durante días')
plt.xlabel('Datos')
plt.ylabel('Duración del sueño (horas)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Plot Movement Count
plt.figure(figsize=(10, 5))
plt.bar(data['Date'], data['Movement_Count'], color='salmon')
plt.title('Movimiento de conteo durante el sueño')
plt.xlabel('Datos')
plt.ylabel('Movimiento de conteo')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Plot Sleep Quality
plt.figure(figsize=(10, 5))
plt.bar(data['Date'], data['Sleep_Quality'], color='lightgreen')
plt.title('Calidad del sueño durante días')
plt.xlabel('Datos')
plt.ylabel('Tasa de la calidad del sueño')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Plot Average Heart Rate
plt.figure(figsize=(10, 5))
plt.bar(data['Date'], data['Heart_Rate_Avg'], color='lightcoral')
plt.title('Tasa Cardíaca Promedio Durante el Sueño Durante Días')
plt.xlabel('Datos')
plt.ylabel('Tasa Cardíaca (BPM)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Plot Caffeine and Alcohol Intake
plt.figure(figsize=(10, 5))
plt.bar(data['Date'], data['Caffeine_Intake'], label='Ingesta de cafeína (mg)', alpha=0.7)
plt.bar(data['Date'], data['Alcohol_Intake'] * 100, label='Ingesta de alcohol (Bebidas Estándar)', alpha=0.5)  # Scale for visibility
plt.title('Ingesta de Cafeína y Alcohol Durante Días')
plt.xlabel('Datos')
plt.ylabel('Ingesta')
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y')
plt.show()
```
